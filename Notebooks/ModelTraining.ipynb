```
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow import keras

import numpy as np
import matplotlib.pyplot as plt
```

```
# Download test set
def getTrainSet(i):
  url = 'https://raw.githubusercontent.com/alexhg128/StocksClassifier/master/TrainSets/train_set_' + str(i).zfill(2) + '.csv'
  return pd.read_csv(url)
```

```
# Creates kera model
def createModel():
  return keras.Sequential([
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(2, activation='softmax')
  ])

# Compiles the model
def compileModel(model):
  model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy', 'categorical_accuracy',
                       'mean_squared_error', 'mean_absolute_error',
                       'mean_absolute_percentage_error'])

# Train a model with a set of values, labels and epochs
def trainModel(model, values, labels, epochs):
  model.fit(values, labels, epochs=epochs)
```

```
# Initialize model
model = createModel()
compileModel(model)
```

```
# For each trainset
for i in range(21):
  # Prepare trainset
  ts = getTrainSet(i + 1)
  ts.pop('Unnamed: 0')
  rec = ts.pop('rec')
  # Train the model
  trainModel(model, ts.values, rec, 15)
```

```
# Print model accuracy
model.history.history['accuracy'][-1]
```

```
# Save model to file
model.save('stocks.h5')
```
